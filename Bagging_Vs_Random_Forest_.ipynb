# Bagging Vs Random Forest:


---

1. Definition

| Feature   | **Bagging**                                    | **Random Forest**                                   |
| --------- | ---------------------------------------------- | --------------------------------------------------- |
| Full Form | Bootstrap Aggregating                          | A specific type of Bagging                          |
| Core Idea | Combines predictions from multiple base models | An ensemble of decision trees with added randomness |


---

2. Base Estimator

| Feature      | **Bagging**                                    | **Random Forest**              |
| ------------ | ---------------------------------------------- | ------------------------------ |
| Base Learner | Can be **any model** (e.g., DecisionTree, SVM) | Always uses **Decision Trees** |


---

3. Feature Selection

| Feature        | **Bagging**           | **Random Forest**                                   |
| -------------- | --------------------- | --------------------------------------------------- |
| Feature Subset | Uses **all features** | Selects **random subset of features** at each split |


---

4. Overfitting

| Feature              | **Bagging**                               | **Random Forest**                                |
| -------------------- | ----------------------------------------- | ------------------------------------------------ |
| Overfitting Tendency | Can still overfit if trees are correlated | Less likely to overfit due to feature randomness |



---

5. Variance Reduction

| Feature                 | **Bagging**                                   | **Random Forest**                                        |
| ----------------------- | --------------------------------------------- | -------------------------------------------------------- |
| How it reduces variance | By averaging predictions from multiple models | Same as Bagging, but more effective due to decorrelation |


---

6. Performance

| Feature    | **Bagging**                                         | **Random Forest**                   |
| ---------- | --------------------------------------------------- | ----------------------------------- |
| Accuracy   | Good, but can be improved                           | Often **better** than plain Bagging |
| Robustness | Sensitive to overfitting if features are correlated | More robust and generalizes better  |


---
7. Out-of-Bag (OOB) Score

| Feature              | **Bagging** | **Random Forest** |
| -------------------- | ----------- | ----------------- |
| OOB Score Supported? | Yes         | Yes               |


---

8. When to Use:

| Use Case                       | Use Bagging                   | Use Random Forest                     |
| ------------------------------ | ----------------------------- | ------------------------------------- |
| Custom base estimators needed  | Yes                           | No (Only Decision Trees)              |
| Higher accuracy and robustness | Not always                    | Preferred in most real-world problems |
| Interpretability               | Better if using simple models | Less interpretable due to many trees  |


---



---

Summary

| Feature               | Bagging | Random Forest        |
| --------------------- | ------- | -------------------- |
| Flexibility in models | Yes     | No                   |
| Reduces variance      | Yes     | Yes (more effective) |
| Avoids overfitting    | No      | Yes                  |
| Feature randomness    | No      | Yes                  |


---



```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf = RandomForestClassifier(max_features=2)
```

```python
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
```

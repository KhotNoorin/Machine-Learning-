# K-Means Clustering Algorithm From Scratch

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans as SKLearnKMeans
from sklearn.metrics import silhouette_score
```

```python
class KMeansScratch:
    def __init__(self, n_clusters=2, max_iter=100, init='random'):
        """
        Parameters:
        ----------
        n_clusters : int
            Number of clusters.
        max_iter : int
            Maximum iterations.
        init : str
            'random' or 'k-means++'
        """
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.init = init
        self.centroids = None

    def fit_predict(self, X):
        """Train and return cluster labels."""
        self._initialize_centroids(X)

        for iteration in range(self.max_iter):
            cluster_labels = self._assign_clusters(X)
            old_centroids = self.centroids.copy()
            self.centroids = self._compute_new_centroids(X, cluster_labels)

            if np.allclose(old_centroids, self.centroids):
                print(f"[Scratch KMeans] Converged in {iteration + 1} iterations.")
                break

        return cluster_labels

    def predict(self, X):
        """Predict cluster for new data points."""
        distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)
        return np.argmin(distances, axis=1)

    def _initialize_centroids(self, X):
        if self.init == 'random':
            indices = random.sample(range(X.shape[0]), self.n_clusters)
            self.centroids = X[indices]
        elif self.init == 'k-means++':
            self.centroids = self._kmeans_plus_plus_init(X)
        else:
            raise ValueError("init must be 'random' or 'k-means++'")

    def _kmeans_plus_plus_init(self, X):
        centroids = []
        centroids.append(X[random.randint(0, X.shape[0] - 1)])

        for _ in range(1, self.n_clusters):
            distances = np.min(np.linalg.norm(X[:, np.newaxis] - np.array(centroids), axis=2), axis=1)
            probs = distances ** 2 / np.sum(distances ** 2)
            cumulative_probs = np.cumsum(probs)
            r = random.random()
            for idx, prob in enumerate(cumulative_probs):
                if r < prob:
                    centroids.append(X[idx])
                    break

        return np.array(centroids)

    def _assign_clusters(self, X):
        distances = np.linalg.norm(X[:, np.newaxis] - self.centroids, axis=2)
        return np.argmin(distances, axis=1)

    def _compute_new_centroids(self, X, cluster_labels):
        new_centroids = []
        for i in range(self.n_clusters):
            points = X[cluster_labels == i]
            if points.size == 0:
                new_centroids.append(self.centroids[i])  # Avoid empty cluster
            else:
                new_centroids.append(points.mean(axis=0))
        return np.array(new_centroids)
```

```python
# Generating some sample data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)
```

```python
# Before clustering visualization
plt.scatter(X[:, 0], X[:, 1], s=30, color='gray')
plt.title("Original Data (Unclustered)")
plt.show()
```

```python
# Using Scratch KMeans (k-means++)
kmeans_scratch = KMeansScratch(n_clusters=4, max_iter=100, init='k-means++')
labels_scratch = kmeans_scratch.fit_predict(X)
```

```python
# Visualize Scratch KMeans
plt.scatter(X[:, 0], X[:, 1], c=labels_scratch, cmap='viridis', s=30)
plt.scatter(kmeans_scratch.centroids[:, 0], kmeans_scratch.centroids[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title("KMeans Scratch Clustering (k-means++)")
plt.legend()
plt.show()
```

```python
# Using scikit-learn KMeans
kmeans_sklearn = SKLearnKMeans(n_clusters=4, init='k-means++', max_iter=100, n_init=10, random_state=42)
labels_sklearn = kmeans_sklearn.fit_predict(X)
```

```python
# Visualize Sklearn KMeans
plt.scatter(X[:, 0], X[:, 1], c=labels_sklearn, cmap='plasma', s=30)
plt.scatter(kmeans_sklearn.cluster_centers_[:, 0], kmeans_sklearn.cluster_centers_[:, 1], c='red', s=200, marker='X', label='Centroids')
plt.title("scikit-learn KMeans Clustering")
plt.legend()
plt.show()
```

```python
# Performance Comparison
scratch_silhouette = silhouette_score(X, labels_scratch)
sklearn_silhouette = silhouette_score(X, labels_sklearn)

print(f"Silhouette Score (Scratch KMeans): {scratch_silhouette:.4f}")
print(f"Silhouette Score (scikit-learn KMeans): {sklearn_silhouette:.4f}")
```

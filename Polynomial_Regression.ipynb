# Polynomial Regression:

---


Polynomial Regression is a type of regression analysis where the relationship between the independent variable x and the dependent variable y is modeled as an n-th degree polynomial.

Polynomial Regression fits a non-linear relationship between the value of x and the corresponding conditional mean of y, denoted as:
y = β0 + β1 x + β2 x2 +⋯+ βn xn +ϵ

---

When to Use:

* When the data shows a non-linear trend.

* When a simple linear regression doesn't provide a good fit.

* Example: Growth curves, pricing models, biological measureme

---

Key Concepts:

* Degree of polynomial: Controls flexibility. Start with low degree and increase.

* Regularization: Use Ridge/Lasso to prevent overfitting in high-degree polynomials.

---

Choosing the Degree of the Polynomial:

Degree 1 → Linear regression

Degree 2 or 3 → Usually enough for many real-world problems

Degree > 5 → May cause overfitting

Use cross-validation to choose the best degree.

---



Advantages:

* Captures non-linear relationships well.

* More flexible than linear regression.


---


Disadvantages:

* Risk of overfitting with high-degree polynomials.

* Sensitive to outliers.

* Extrapolation can be misleading.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression,SGDRegressor

from sklearn.preprocessing import PolynomialFeatures,StandardScaler

from sklearn.metrics import r2_score

from sklearn.pipeline import Pipeline
```

```python
X = 6 * np.random.rand(200, 1) - 3
y = 0.8 * X**2 + 0.9 * X + 2 + np.random.randn(200, 1)

# y = 0.8x^2 + 0.9x + 2
```

```python
plt.plot(X, y,'b.')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# Train test split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Applying linear regression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train,lr.predict(X_train),color='r')
plt.plot(X, y, "b.")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

# Applying Polynomial Linear Regression (degree 2)

```python
poly = PolynomialFeatures(degree=2,include_bias=True)

X_train_trans = poly.fit_transform(X_train)
X_test_trans = poly.transform(X_test)
```

```python
print(X_train[0])
print(X_train_trans[0])
```

```python
# include_bias parameter
lr = LinearRegression()
lr.fit(X_train_trans,y_train)
```

```python
y_pred = lr.predict(X_test_trans)
```

```python
r2_score(y_test,y_pred)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
X_new=np.linspace(-3, 3, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
```

```python
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(X_train, y_train, "b.",label='Training points')
plt.plot(X_test, y_test, "g.",label='Testing points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
def polynomial_regression(degree):
    X_new=np.linspace(-3, 3, 100).reshape(100, 1)
    X_new_poly = poly.transform(X_new)

    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    polynomial_regression = Pipeline([
            ("poly_features", polybig_features),
            ("std_scaler", std_scaler),
            ("lin_reg", lin_reg),
        ])
    polynomial_regression.fit(X, y)
    y_newbig = polynomial_regression.predict(X_new)
    plt.plot(X_new, y_newbig,'r', label="Degree " + str(degree), linewidth=2)

    plt.plot(X_train, y_train, "b.", linewidth=3)
    plt.plot(X_test, y_test, "g.", linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-3, 3, 0, 10])
    plt.show()
```

```python
polynomial_regression(50)
```

```python
# Applying Gradient Descent

poly = PolynomialFeatures(degree=2)

X_train_trans = poly.fit_transform(X_train)
X_test_trans = poly.transform(X_test)

sgd = SGDRegressor(max_iter=100)
sgd.fit(X_train_trans,y_train)

X_new=np.linspace(-2.9, 2.8, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
y_new = sgd.predict(X_new_poly)

y_pred = sgd.predict(X_test_trans)

plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions " + str(round(r2_score(y_test,y_pred),2)))
plt.plot(X_train, y_train, "b.",label='Training points')
plt.plot(X_test, y_test, "g.",label='Testing points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

üîπ Assumptions of Linear Regression:
Linearity between independent and dependent variables

Homoscedasticity (constant variance of errors)

Independence of errors

No multicollinearity (in multiple regression)

Normal distribution of errors



How It Works


1.   Hypothesis: Predicts y using a linear equation.
2.   Loss Function: Measures error between predicted and actual values using Mean Squared Error (MSE):
3.  Optimization: Uses Gradient Descent or analytical solutions (like the Normal Equation) to minimize the MSE.






Limitations of Linear Regression


1.   Assumes Linear Relationship
2.   Sensitive to Outliers
3.   Multicollinearity
4.   Assumes Homoscedasticity (assumes that error variance is constant across all levels of input.)
5.    Assumes Independence of Errors
6.    No Automatic Feature Selection
7.    Not Great for Complex Problems
8.    Needs Numerical Inputs
9. Assumes Normality of Errors




```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([2,4,6,8,10])
```

```python
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
y_pred=model.predict(x)
```

```python
print("Intercept:", model.intercept_)  # constant term 'b'
print("Slope:", model.coef_[0])        # weight 'm' in y = mx + b
```

```python
plt.scatter(x,y,color='Blue',label='Actual')
plt.plot(x,y_pred,color='Red',label="Predicted")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x1=np.array([
    [1000, 2],
    [2000, 6],
    [4000, 5],
    [9000, 7],
    [6000, 6],
    [4000, 4]
])
```

```python
y1=np.array([200,300,400,700,800,900])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("Predicted Prices:", y_pred)
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
print("Model Score (R¬≤):", model.score(x_test, y_test))
```

 Manual Multiple Linear Regression

```python
import numpy as np


X = np.array([
    [1, 2, 3],
    [2, 3, 4],
    [3, 4, 5],
    [4, 5, 6],
    [5, 6, 7]
], dtype=float)

y = np.array([10, 13, 16, 19, 22], dtype=float)

#Initialize parameters
n_samples, n_features = X.shape
weights = np.zeros(n_features)  # weights for each feature
bias = 0.0


alpha = 0.01
epochs = 1000

#Gradient Descent Loop
for i in range(epochs):
    y_pred = np.dot(X, weights) + bias     # predicted y
    error = y - y_pred                     # residuals

    # Compute gradients
    dw = (-2 / n_samples) * np.dot(X.T, error)
    db = (-2 / n_samples) * np.sum(error)

    # Update parameters
    weights -= alpha * dw
    bias -= alpha * db

    # Print loss every 100 epochs
    if i % 100 == 0:
        mse = np.mean(error ** 2)
        print(f"Epoch {i}: Loss = {mse:.4f}, Weights = {weights}, Bias = {bias:.4f}")

# model
print("\nFinal Weights:", weights)
print("Final Bias:", bias)
```

```python
# Example prediction
x_test = np.array([6, 7, 8])  # new input
y_test_pred = np.dot(x_test, weights) + bias
print("\nPrediction for input [6, 7, 8]:", y_test_pred)
```

Accuracy measure in linear regression:

Mean Squared Error (MSE)

Root Mean Squared Error (RMSE)

Mean Absolute Error (MAE)

R¬≤ Score

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_true = np.array([2.2, 2.8, 4.5, 3.7, 5.5])
y_pred = np.array([2.0, 3.0, 4.0, 4.0, 5.0])

mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print(f"MSE: {mse:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"MAE: {mae:.3f}")
print(f"R¬≤ Score: {r2:.3f}")
```

Activation Functions

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # for numerical stability
    return exp_x / np.sum(exp_x)
```
